```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
TRAIN_DF = "../../../data/training_set.csv"
```

```python
df_train = pd.read_csv(TRAIN_DF)
```

```python
df_train.head()
```

```python
df_train.describe()
```

```python
df_train[(df_train.object_id == 615) & (df_train.passband == 0)].plot.scatter(x="mjd", y="flux", figsize=(16,8))
```

```python
df_train[df_train.object_id == 615].plot.scatter(x="mjd", y="flux", c="passband", colormap="viridis", figsize=(16,8))
```

```python
sample_obj = df_train[df_train.object_id == 615]
```

```python
sns.set(style="ticks")
fig, ax = plt.subplots(figsize=(16,8))
fontdict = {'fontsize': 14}
ax.set_title(label="Boxplots of flux per passband for the object_id 615", fontdict=fontdict, loc='center', pad=None)
sns.boxplot(x=sample_obj.passband, y=sample_obj.flux, hue=sample_obj.passband, palette=sns.color_palette("coolwarm", 6))
```
